```python
# Google Drive Source Data
# https://docs.google.com/spreadsheets/d/1BN5nlzQRauA_xnpLds3fDw4GBquiyd8A/edit?usp=drive_link&ouid=102514651699712419557&rtpof=true&sd=true
```

```python
url = 'https://docs.google.com/spreadsheets/d/1BN5nlzQRauA_xnpLds3fDw4GBquiyd8A/edit?usp=drive_link&ouid=102514651699712419557&rtpof=true&sd=true'
file_id = url.split('/')[-2]
read_url='https://drive.google.com/uc?id=' + file_id

# Level Granularity:
# -  Customer ID
#  -  Order_ID
#   -  Category
#    -  Sub-Category
#     -  Product_ID - Product_Name
```

```python
# read the data
ss = pd.read_excel(read_url)
ss.head()
```

# **Summary Data**

```python
sum_category = pd.pivot_table(
  data=ss,
  index='Category',
  values='Sales',
  aggfunc='sum'
)
sum_category
```

```python
sum_category = pd.pivot_table(
  data=ss,
  index='Category',
  values='Sales',
  aggfunc='sum'
)
sum_category
```

```python
sum_cat_sku = pd.pivot_table(
  data=ss,
  index= ['Category', 'Product_Name'],
  values='Sales',
  aggfunc='sum'
)
sum_cat_sku
```

```python
sum_cat_sku = pd.pivot_table(
  data=ss,
  index= ['Category', 'Product_Name', 'Sub-Category'],
  values='Sales',
  aggfunc='sum'
)
sum_cat_sku
```

```python
sum_cat_sku = pd.pivot_table(
  data=ss,
  index= ['Category', 'Product_Name', 'Sub-Category'],
  values='Sales',
  aggfunc= ['sum', 'mean']
)
sum_cat_sku
```
