## MULTIPLE CONDITIONS

```python
# memfilter dengan dua atau lebih kondisi
# ingat : untuk menghindari kesalahan penulisan ekspresi, gunakan tanda kurung

# gunakan bitwise operator
# & = and
# | = or
# ~ = not

# memfilter transaksi dengan sales > 500



df[(df['sales'] > 500) & (df['quantity'] > 10)].head()
```

```python
df.query('sales > 500 & quantity > 10').head()
```

```python
df[(df['sales'] <= 500) & (df['quantity'] <= 10)].head()
```

### ~ (NOT)

```python
df[~(df['category'] == 'Furniture')].head()
```

# SUMMARIZING DATA

```python
df.groupby(['order_date', 'region', 'category']).agg({'sales' : 'sum', 'quantity' : 'sum'})
```

```python
summarize_df = df.groupby(['order_date', 'region', 'category']).agg({'sales' : 'sum', 'quantity' : 'sum'})
summarize_df
```

```python
# sales, quantity di identify sbg kolom
# order_date, region, category di indentify sebagai index

summarize_df[['sales', 'quantity']]
```

```python
# Hanya preview, belum ter-save

summarize_df.reset_index()
```
