# **HANDLING MISSING VALUES**

```python
# 2 metode handling missing value:

df_delete = df.copy()
df_impute = df.copy()
```

## **DELETE**

```python
# menghapus row yang kolom age-nya kosong
df_delete = df_delete.dropna(subset=["Age"], axis = 0)
df_delete
```

```python
df_delete.info()
```

## **IMPUTE**

```python
df_impute_series = df_impute.copy()
df_impute_partial = df_impute.copy()
```

## df_impute_series

```python
df_impute_series.describe()
```

```python
df_impute ["Age"].plot(kind="density")
```

```python
from scipy import stats
res= stats.normaltest(df_impute_series['Age'].values, nan_policy="omit")
print(res.pvalue)
```

*kenapa hasilnya masih beda ya?*

```python
df_impute ['Age'] = df_impute_series ['Age'].fillna(df_impute_series ['Age'].mean())
df_impute_series
```

## df_impute_partial

```python
for i in ['male', 'female']:
  print(f"{i} age")
  res = stats.normaltest(df_impute_partial[df_impute_partial['Sex']==i]['Age'], nan_policy="omit")
  p = res.pvalue
  print(p)
```

```python
male_age = df_impute_partial[df_impute_partial['Sex']=='male']['Age'].fillna(df_impute_partial[df_impute_partial['Sex']=='male']['Age'].mean())
female_age = df_impute_partial[df_impute_partial['Sex']=='female']['Age'].fillna(df_impute_partial[df_impute_partial['Sex']=='male']['Age'].median())
```
